# NutriPlan AI Benchmarks

This notebook loads the saved benchmark, security, and timing artifacts for review.


```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt

BASE_DIR = Path.cwd()
if not (BASE_DIR / "benchmark_summary.csv").exists():
    BASE_DIR = Path("benchmarks")
print("Using results from:", BASE_DIR.resolve())
```

## Model Comparison Summary


```python
summary = pd.read_csv(BASE_DIR / "benchmark_summary.csv")
display(summary)
```

```python
ax = summary.plot(kind="bar", x="model", y="avg_latency_seconds", legend=False, figsize=(7,4), color="#2f9e44")
ax.set_ylabel("Average latency (seconds)")
ax.set_xlabel("")
ax.set_title("Average Response Latency")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()
```

## Query-Level Results


```python
results = pd.read_csv(BASE_DIR / "benchmark_results.csv")
display(results[["model_key", "query_id", "latency_seconds", "response_length", "num_tool_calls", "error"]].head(10))
display(results.groupby("model_key")[["latency_seconds", "response_length", "num_tool_calls"]].mean().round(3))
```

## Security Results


```python
security = pd.read_csv(BASE_DIR / "security_results.csv")
display(security[["model_key", "attack_id", "verdict", "error"]])
display(security.groupby(["model_key", "verdict"]).size().unstack(fill_value=0))
```
